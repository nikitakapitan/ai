```python
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
def foo(x):
    return 3*np.sin(0.8*x) + np.random.rand()
```

```python
xmin, xmax = (1, 8)
nb = 30

xx = np.linspace(xmin, xmax, nb) 
yy = np.array([foo(x) for x in xx])
degrees = [0,1,2,3, 50]
```

```python
p = {}
for deg in degrees:
    p[deg] = np.poly1d(np.polyfit(xx, yy, deg))
```

### Seems like deg=2 is the best non overfitted model

```python
fig= plt.figure(figsize=(10,8))

plt.scatter(xx, yy);
for deg in degrees:
    plt.plot(xx, p[deg](xx), label = f'deg={deg}');
plt.legend();
```

## But for extrapolation the best estimator is constant :D

```python
xxp = np.linspace(xmax, 2*xmax, nb) 
xx2 = np.concatenate((xx, xxp))
yy2 = np.concatenate((yy, np.array([foo(x) for x in xxp])))


fig= plt.figure(figsize=(20,8))
plt.scatter(xx2, yy2);
for deg in degrees:
    plt.plot(xx2, p[deg](xx2), label = f'deg={deg}');
plt.legend();
plt.ylim((-4, 6));
```
